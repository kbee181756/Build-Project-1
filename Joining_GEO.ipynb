```
from geopandas import GeoDataFrame
```

```
from shapely.geometry import Point
```

```
long_lat.head()
```

```
geometry = [Point(xy) for xy in zip(long_lat.LONGITUDE, long_lat.LATITUDE)]
```

```
geometry
```

```
long_lat = long_lat.drop(['LONGITUDE', 'LATITUDE'], axis=1)
```

```
crs = {'init': 'epsg:4326'}
```

```
gdf = GeoDataFrame(long_lat, crs=crs, geometry=geometry)
```

```
gdf
```

```
gpd.sjoin(left_df=nycdoe, right_df=gdf, how='left')
```
